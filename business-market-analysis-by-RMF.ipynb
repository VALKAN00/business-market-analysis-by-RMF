```python
# Reference date for recency
last_day = df2['Date'].max() + dt.timedelta(days=1)

# Calculate RFM metrics
rfm = df2.groupby('Customer_ID').agg({'Date': lambda x: (last_day-x.max()).days, 
                                        'Transaction_ID':'count', 
                                       'Sales_Amount':'sum'})
rfm.rename(columns={'Date':'recency','Transaction_ID':'frequency','Sales_Amount':'monetary_value'}, inplace=True)
rfm.head(10)
```

```python
r_quartile = range(3,0,-1)
f_quartile = range(1,4) 
m_quartile = range(1,4)
r_segments = pd.qcut(rfm['recency'], q=3, labels= r_quartile)
f_segments = pd.qcut(rfm['frequency'], q=3, labels= f_quartile)
m_segments = pd.qcut(rfm['monetary_value'], q=3, labels= m_quartile)
rfm = rfm.assign(r = r_segments, f=f_segments, m=m_segments)

#Calculating the RFM segment and RFM score for each customer
rfm['segment'] = rfm.apply(lambda x: str(int(x['r']))+str(int(x['f']))+str(int(x['m'])), axis=1)
rfm['score'] = rfm[['r','f','m']].sum(axis=1)

rfm.head(10)
```
